# First steps

## Метод конечных разностей - FDM
Есть два метода решения уравнений - аналитический и численный. Для аналитического нужно хорошо понимать математику. Для численного нет (это +). Так же только численные методы могут решать задачу в общем случае. FDM - finite-difference method (метод конечных разностей).

## Производная
![Картинка с производной](https://upload.wikimedia.org/wikipedia/commons/d/d2/Tangent-calculus.svg)

Производная - это угол наклона касательной к графикку в данной точке(с точки зрения анлитического подхода). 
Метод конечных разностей рассматривает производную просто как разность двух соседних известных значений.

![image for fdm](http://i63.tinypic.com/2eowld0.jpg)

В этом случае мы определяем производную слежующим образом:

$$\frac{\partial \Phi}{\partial x} = \frac{\Phi_1 - \Phi_2}{x_1 - x_2} (1)$$

Где фи - какое то значение по оси Y а x какое то значение по оси X

## Производные больших порядков

По аналогии для больших для производных больших порядков действует тот же принцип - разница между соседними известными значениями:

![image for hight order derivation]()

Что можно записать так:

$$ \frac{\partial^2 \Phi}{\partial x^2} = \frac{\left(\frac{\partial \Phi}{\partial x}\right)_B - \left(\frac{\partial \Phi}{\partial x}\right)_A}{x_B - x_A} $$

## Prons and cons

Основное достоинство FDM - оно позволяет уйти от бесконечных аналитических расчетов и считать уравнения в частных производных в конкретных точках. Обобщенный алгоритм применения конечных разностей состоит из следующих шагов: 
* Создать сетку на которой будет происходить расчет ур-я
* Определить физические характеристики этих точек( то есть что мы будем в них считать)
* Учесть граничные условия
* Применить FDM и получить систему линейных уравнений
* Решить систему линейных уравнений и получить неизвестные физические свойства точек

## Решение систем линейных уравнений

$$\begin{bmatrix}
L1,1S1 + L1,2S2 + L1,3S3 +···+ L1,n−1Sn−1 + L1,nSn = R1 \\
L2,1S1 + L2,2S2 + L2,3S3 +···+ L2,n−1Sn−1 + L2,nSn = R2 \\
    ... \\
Ln−1,1S1 + Ln−1,2S2 + Ln−1,3S3 +···+ Ln−1,n−1Sn−1 + Ln−1,nSn = Rn−1 \\
Ln,1S1 + Ln,2S2 + Ln,3S3 +···+ Ln,n−1Sn−1 + Ln,nSn = Rn
\end{bmatrix}$$

Где S - неизвестные, L - коэффициенты, R - известные правой части. Системы линейных уравнений могут быть решены только при условии равенства количества уравнений и количества неизвестных.

В сулучае численных расчетов большинство коеффициентов L равны 0, так как в уравнении содержатся члены только для соседних узлов сетки. Таким образом решение уравнений сводится к правильному определению коэффициентов L и правой части R и последующее автоматическое решение СЛАУ.
